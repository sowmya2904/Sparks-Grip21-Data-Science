## Data Visualization

#### Distribution of numerical data

```python
fig, axs=plt.subplots(ncols=2,nrows=2,figsize=(10, 10))
sns.distplot(df['Profit'], color='green',ax=axs[0][0])
sns.distplot(df['Sales'], color='red',ax=axs[0][1])
sns.distplot(df['Discount'], color='orange',ax=axs[1][0])
sns.distplot(df['Quantity'], color='blue',ax=axs[1][1])
axs[0][0].set_title('Profit Distribution',fontsize=20)
axs[0][1].set_title('Sales Distribution',fontsize=20)
axs[1][0].set_title('Discount Distribution',fontsize=20)
axs[1][1].set_title('Quantity Distribution',fontsize=20)
plt.show()
```

<font size="4"> From above we can see that there is no normal distribution in numerical data except Profit </font>

#### Check for correlation in data

```python
#Check for the correlation of the data using heatmap
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),linecolor='Black',linewidth=10,annot=True);
```

<font size="4"> Discount & Profit are highly negatively correlated and Sales & Profit are highly postively correlated </font>

```python
#Pattern of correlation of Discount and Profit
plt.scatter(x=df['Discount'], y=df['Profit'], color='crimson', alpha=0.4)
plt.title('Discount/Profit', weight='bold', fontsize=16)
plt.xlabel('Discount', weight='bold', fontsize=12)
plt.ylabel('Profit', weight='bold', fontsize=12)
plt.xticks(weight='bold')
plt.yticks(weight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.lineplot('Discount','Profit', data=df , color='y',label='Discount')
plt.legend()
plt.show()
```

<font size="4"> When Discount is 0, then profit is high. As discount increases, Profit is decreasing till 0.5. When dicount is 0.6, then profit has increased drastically, and after that slowly decreasing as discount is increasing. </font>
    

```python
# Regression Plot 
sns.regplot(df['Sales'],df['Profit'])
```

<font size="4"> There is positive linearity betweeen Sales and Profit </font>

### Lets explore categorical data

```python
# extract those categorical columns 
categorical_col = []
for column in df.columns:
    if df[column].dtypes != "float64" and df[column].dtypes != "int64" and df[column].dtypes != "bool":
        categorical_col.append(column)
```

```python
categorical_col
```

```python
# Extract Top 10 cities where sales and profit are high
Top_10_Sales = df.groupby("City").Sales.sum().nlargest(n =10)
Top_10_Profits = df.groupby("City").Profit.sum().nlargest(n =10)
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20,5))
plt.style.use('seaborn')
Top_10_Sales.plot(kind ='bar', figsize =(14,8), fontsize =14,ax=ax1)
Top_10_Profits.plot(kind ='bar', figsize =(14,8), fontsize =14,ax=ax2)
ax1.set_title('Top 10 Cities By Sales',fontsize=20)
ax2.set_title(' Top 10 Cities By Profits',fontsize=20)
plt.show()
```

- New York City, Los Angeles, Seattle and San Francisco have highest sales and profits both.
- San Diego have less profit compared to sales
- Cities like, Detroit, Lafayette, Jackson, Atlanta and Mineneapolis does not come under top 10 sales. But still they have large profits.
